```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[["Profit"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='random',max_features = 'log2')
regressor=regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
def decisionTreeRegressor_rscore(crt,spl,feat):    
    regressor=DecisionTreeRegressor(criterion=crt, splitter=spl,max_features = feat)
    regressor=regressor.fit(X_train,y_train)
    y_pred=regressor.predict(X_test)   
    r_score=r2_score(y_test,y_pred)
    return r_score
```

```python
import pandas as pd
algorithim_name =  "DecisionTreeRegressor"
criterion_list = ['squared_error','friedman_mse','absolute_error','poisson']
splitter_list = ['best','random']
max_features_list = ['sqrt','log2']
final_df = pd.DataFrame(columns=['Algorithim','Criterion','Splitter','Max_features','R-Score'])

for crt in criterion_list:
    for spl in splitter_list:
        for feat in max_features_list:
            r_scr = decisionTreeRegressor_rscore(crt,spl,feat)
            data_to_store = {'Algorithim' : "DecisionTreeRegressor",'Criterion' : crt ,'Splitter' : spl ,'Max_features' : feat ,'R-Score' : r_scr}
            final_df.loc[len(final_df)] = data_to_store 
print(final_df)
```
